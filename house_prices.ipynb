```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Dataset Loading
df = pd.read_csv("house_prices.csv")  
df.head()
```

```python
#Identifying missing values using .isnull().sum()
df.isnull().sum()
```

```python
#Visualize missing data patterns using simple bar charts
df.isnull().sum().plot(kind='bar', figsize=(10,4))
plt.title("Missing Values Per Column")
plt.show()
```

```python
#.Applying median imputation for numerical columns
num_cols = df.select_dtypes(include=['int64','float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
#Applying mode imputation for categorical columns
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
#Remove columns with extremely high missing values.
threshold = 0.5
df = df.loc[:, df.isnull().mean() < threshold]
```
